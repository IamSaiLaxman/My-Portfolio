```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load data
file_path = '/content/Global_Cybersecurity_Threats_2015-2024.csv'
df = pd.read_csv(file_path)

# Step 2: Rename columns for easier handling (if needed)
df.columns = df.columns.str.strip()

# Step 3: Select relevant columns
cols = [
     'Attack Type', 'Target Industry',
   'Incident Resolution Time (in Hours)'
]

df = df[cols].dropna()

# Step 4: Separate features by type
numerical = [ 'Incident Resolution Time (in Hours)']
categorical = ['Attack Type', 'Target Industry']

# Step 5: Preprocess (scale numeric, one-hot encode categorical)
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical)
])

X = preprocessor.fit_transform(df)

# Step 6: Elbow method to choose K
inertia = []
for k in range(1, 8):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X)
    inertia.append(km.inertia_)

plt.plot(range(1, 8), inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

# Step 7: Fit KMeans with chosen K (e.g., K=3)
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Step 8: Visualize with PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=df['Cluster'], palette='Set2')
plt.title('Multivariate Clustering (PCA)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()

# Step 9: Cluster summary
summary = df.groupby('Cluster').agg({
    'Incident Resolution Time (in Hours)': ['mean', 'median'],


    'Attack Type': lambda x: x.mode()[0],
    'Target Industry': lambda x: x.mode()[0],

})

print("Cluster Summary:\n")
print(summary)
```

```python
df = pd.read_csv('/content/Global_Cybersecurity_Threats_2015-2024.csv')
# Binning Incident Resolution Time into Fast, Medium, Slow
def categorize_resolution_time(hours):
    if hours <= 12:
        return 'Fast'
    elif hours <= 48:
        return 'Medium'
    else:
        return 'Slow'

df['Resolution_Category'] = df['Incident Resolution Time (in Hours)'].apply(categorize_resolution_time)

# Encode categorical columns
label_cols = ['Country', 'Attack Type', 'Target Industry', 'Attack Source',
              'Security Vulnerability Type', 'Defense Mechanism Used']

le = LabelEncoder()
for col in label_cols:
    df[col] = le.fit_transform(df[col])

# Drop string-based or redundant columns if they exist
if 'Financial Loss (in Million $)' in df.columns and df['Financial Loss (in Million $)'].dtype == 'O':
    df = df.drop(columns=['Financial Loss (in Million $)'])

# Features and Target
X = df.drop(columns=['Incident Resolution Time (in Hours)', 'Resolution_Category'])
y = df['Resolution_Category']

# Scale numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y, test_size=0.2, random_state=42)

# Train model
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Predict
y_pred = clf.predict(X_test)

# Evaluate
print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=clf.classes_, yticklabels=clf.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
